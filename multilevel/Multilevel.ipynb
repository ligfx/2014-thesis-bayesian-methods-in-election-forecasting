```python
%matplotlib inline
%load_ext rmagic

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
```

```python
srrs2 = pd.read_csv("srrs2.dat")

srrs2.columns = (c.strip() for c in srrs2.columns)
srrs2 = srrs2[srrs2['state'] == 'MN']

#nure = pd.read_csv("cty.dat")
#nure = nure[nure['st'] == 'MN']

def normalize_county(cty):
	return cty.lower().replace(" ", "")

#nure['cty'] = nure['cty'].map(normalize_county)
srrs2['county'] = srrs2['county'].map(normalize_county)

# Dropping the items that can't decide whether they're in a basement or not
srrs2 = srrs2[(srrs2['floor'] == 0) | (srrs2['floor'] == 1)]
# And ones with absolutely 0 activity. It might be better to just make this very small?
srrs2 = srrs2[srrs2['activity'] > 0]
# Logradon
srrs2['logradon'] = np.log(srrs2['activity'])
```

```python
pooled = smf.ols('logradon ~ floor', data=srrs2).fit()
unpooled = smf.ols('logradon ~ floor + county - 1', data=srrs2).fit()
```

```python
%%R -i srrs2 -o partial_coef -o partial_labels

library('lme4')

# pooled <- lm(logradon ~ floor, srrs2)
# summary(pooled)
# unpooled <- lm(logradon ~ floor + factor(county) - 1, srrs2)
# summary(unpooled)

partial <- lmer(logradon ~ floor + (1 | county), srrs2)
partial_coef <- coef(partial)$county
partial_labels <- labels(partial_coef)[1]
```

```python
partial = dict(zip(partial_labels[0], zip(*partial_coef)))

def partial_predict(county, floor):    
    return partial[county][0] + partial[county][1] * floor
```

```python
srrs2['jitter'] = np.random.normal(0, 0.01, len(srrs2))

plt.figure()
plt.grid()
plt.scatter(srrs2['floor'] + srrs2['jitter'], srrs2['logradon'])
plt.plot([-0.1, 1.1], [pooled.predict({'floor': -0.1}), pooled.predict({'floor': 1.1})], 'b--')
plt.xlim([-0.1, 1.1])
plt.ylim([-1.5, 3.5])
```

```python
for countyname in ['lacquiparle', 'stlouis']:
    county = srrs2[srrs2['county'] == countyname]

    plt.figure()
    plt.grid()
    plt.scatter(county['floor'] + county['jitter'], county['logradon'])
    plt.plot([-0.1, 1.1], [pooled.predict({'floor': -0.1}), pooled.predict({'floor': 1.1})], 'b--')
    plt.plot([-0.1, 1.1], [
        unpooled.predict({'county': [countyname], 'floor': -0.1}),
        unpooled.predict({'county': [countyname], 'floor': 1.1})], 'r')
    plt.plot([-0.1, 1.1], [partial_predict(countyname, -0.1), partial_predict(countyname, 1.1)], 'k')
    plt.title("hi")
    plt.xlim([-0.1, 1.1])
    plt.ylim([-1.5, 3.5])
```
